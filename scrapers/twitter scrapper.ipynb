<h1>Tweet Scraper</h1>

This notebook uses **Tweepy Twitter scrapper** in order to scrape data from twitter.
**Please notice:** Twitter is making it pretty hard to scraped multiple tweets from thier site.
**You are limited to 1000 observations per scraping. Also, You might not be able to run all this notebok at once, because twitter will blok you.**
It's better to scrap 2 line at a time, and than restart the karnel. That is why I split it this way.
Also, notice that you are not puting to much key words in the searching line.
