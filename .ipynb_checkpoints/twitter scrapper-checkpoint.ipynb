<h1>Tweet Scraper</h1>

This notebook uses **Tweepy Twitter scrapper** in order to scrape data from twitter.
**Please notice:** Twitter is making it pretty hard to scraped multiple tweets from thier site.
**You are limited to 1000 observations per scraping. Also, You might not be able to run all this notebok at once, because twitter will blok you.**
It's better to scrap 2 line at a time, and than restart the karnel. That is why I split it this way.
Also, notice that you are not puting to much key words in the searching line.

```python
neg_line1 = "#IDoNotConsent OR #IWillNOTComply OR OR #Idonotconsent OR #novaccine OR #idonotconsen OR #AntiVacciners"
neg_line2 = "antivaxx OR #antivaxxers OR #Antivaxx OR #TrumpsVaccineIsALie OR #antivax OR #CovidHoax OR #antivaxx OR #antivaxers"
```

```python
import tweepy as tw
import pandas as pd
import datetime

auth = tw.OAuthHandler('d5HZe4mtTtZDAJllEnbuyaT9v', 'mY7Jlf9r1j3BOI20QSB84T3ktByMxjwjAhL6rrRRjrBckBShE0')
auth.set_access_token('230756541-J0P9NtPULFfEebC5mOQoB1k4n6dJwd7yk8NPGjHd', 'DY0UlIsKTG6FX7AuQVUCTQx248Mbta8J76FuSyHTMezDB')
api = tw.API(auth)
date_since = "2020-01-01"
date_until= "2021-01-10"
```

```python
search_words = neg_line1

tweets = tw.Cursor(api.search,
              q=search_words,
              lang="en",
              since=date_since).items(1000)
content = []
id = []
username = []
location = []
description = []
user_creat_date = []
followers = []
favorite_tweet_count = []
retweet_count = []
posted_at = []
is_retweet = []

for i, tweet in enumerate(tweets):
    print(i)
    content.append(tweet.text)
    id.append(tweet.id)
    username.append(tweet.user.screen_name)
    location.append(tweet.user.location)
    description.append(tweet.user.description)
    user_creat_date.append(tweet.user.created_at)
    followers.append(tweet.user.followers_count)
    favorite_tweet_count.append(tweet.favorite_count)
    retweet_count.append(tweet.retweet_count)
    posted_at.append(tweet.created_at)
    is_retweet.append(tweet.retweet_count > 0)
    
data = {'id': id,
        'user_name': username,
        'user_location':location,
        'user_description':description,
        'user_created': user_creat_date,
        'user_followers':followers,
        'date':posted_at,
        'text':content,
        'favorites':favorite_tweet_count,
        'retweets':retweet_count,
        'is_retweet':is_retweet
        }

neg1 = pd.DataFrame(data)
print(neg1.shape)
neg1.head()
```
